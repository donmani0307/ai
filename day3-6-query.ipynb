```python
long_exam = pd.read_csv("Data2/exam.csv")
```

```python
long_exam.assign(new = long_exam['math'] + long_exam['english'] + long_exam['science'])
```

```python
# lambda 사용해서 long_name 대신에 x 사용
long_exam.assign(new = lambda x : x['math'] + x['english'] + x['science'])
```

```python
# lambda를 사용하여 변수 재활용
exam.assign(total = exam['math'] + exam['english'] + exam['science'],
            mean = lambda x : x['total'] / 3)
```

```python
# 집단 별로 요약하기
```

```python
# 수학점수 평균
exam.agg(mean_math = ('math', 'mean'))
```

```python
# 반별로 수학평균
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))
```

```python
# 변수를 인덱스로 바꾸지 않게 하기
exam.groupby('nclass', as_index=False) \
    .agg(mean_math = ('math', 'mean'))
```

```python
mpg = pd.read_csv("Data2/mpg.csv")
mpg
```

```python
# 제조회사 및 구동 방식별로 분리
# cty 평균 구하기
```

```python
mpg.groupby(['manufacturer', 'drv']) \
    .agg(mean_cty = ('cty', 'mean'))
```

```python
# audi의 drv별 빈도
mpg.query('manufacturer == "audi"') \
    .groupby(['drv']) \
    .agg(n = ('drv', 'count'))
```
